```python
# import the cancer dataset from scikit-learn, split it into training and test sets, and then train a random forest classifier
# on the dataset. Finally, we evaluate the classifier on the test set.
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load the dataset
digits_df = load_digits(as_frame=True)

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(digits_df.data, digits_df.target, random_state=0)
X_train
```

```python
y_train
```

```python
# Create a random forest classifier
clf = RandomForestClassifier(n_estimators=10, random_state=42)

# Train the classifier
clf.fit(X_train, y_train)

# Print the name and gini importance of each feature
for feature in zip(digits_df.feature_names, clf.feature_importances_):
    print(feature)
```

```python
# Apply the classifier we trained to the test data (which, remember, it has never seen before)
y_pred = clf.predict(X_test)

# Evaluate the classifier on the test data
acc1 = accuracy_score(y_test, y_pred)
```

```python
# compare the performance of the random forest classifier to that of a gradient boosted tree classifier

from sklearn.ensemble import GradientBoostingClassifier

# Create a gradient boosted tree classifier

clf2 = GradientBoostingClassifier(random_state=0)

# Train the classifier
clf2.fit(X_train, y_train)
```

```python
# Apply the classifier we trained to the test data (which, remember, it has never seen before)
y_pred = clf2.predict(X_test)

# Evaluate the classifier on the test data
acc2 = accuracy_score(y_test, y_pred)
```

```python
# compare the performance of the random forest classifier to that of a histogram gradient boosted tree classifier

from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier

# Create a histogram-based gradient boosted tree classifier
clf3 = HistGradientBoostingClassifier(random_state=0)

# Train the classifier
clf3.fit(X_train, y_train)

# Apply the classifier we trained to the test data (which, remember, it has never seen before)
y_pred = clf3.predict(X_test)

# Evaluate the classifier on the test data
acc3 = accuracy_score(y_test, y_pred)
```

```python
# create a function that plots the obtained accuracy scores for the three classifiers

import matplotlib.pyplot as plt
import numpy as np

def plot_feature_importances_cancer(model):
    n_features = digits_df.data.shape[1]
    plt.barh(np.arange(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), digits_df.feature_names)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    plt.ylim(-1, n_features)

plot_feature_importances_cancer(clf)
```

```python
# plot the accuracy scores for the three classifiers
plt.barh(range(3), [acc1, acc2, acc3])
```

```python
# plot the accuracy scores for the three classifiers and add labels
plt.barh(range(3), [acc1, acc2, acc3])
plt.yticks(range(3), ['Random Forest', 'Gradient Boosted', 'Histogram-based Gradient Boosting'])
plt.xlabel('Accuracy')
plt.xlim(0.8, 1.0)
plt.show()
```
